```python
#2a
def is_prime (n):
  if n < 2:
    return False
  if n == 2:
    return True
  if n % 2 == 0:
    return False    
  for i in range(3,int(n**0.5),2):
    if n % i == 0:
      return False
  return True
```

```python
is_prime(7)
```

```python
is_prime(7)
```

```python
#2b
def sum_from_1_to_N(n):
  sum = 0
  for i in range (n+1):
    sum+=i
  return sum
```

```python
sum_from_1_to_N(10)
```

```python
#2c
def prime_from_1_to_N(n):
  primelist = []
  for i in range(n+1):
    if (is_prime(i)):
      primelist.append(i)
  return primelist
```

```python
prime_from_1_to_N(10)
```

```python
#2d

def n_prime_numb(n):
  primelist = []
  currentNumb = 2
  while (len(primelist)<n):
      if (is_prime(currentNumb)):
          primelist.append(currentNumb)
      currentNumb+=1
  return primelist
      

```

```python
n_prime_numb(10)
```

```python
#3a

def sumNumbList (a):
    sum_ = 0
    for i in range (len(a)):
        sum_+=a[i]
    return sum_
```

```python
#3c
def add_member (array,n):
    array.append(n)
    return array
    
```

```python
#3d
def delete_member (index,array):
    if index >= 0 and index < len(array)+1:
        array.pop(index)
    return array
```

```python
array = [1,5,8]

delete_member (0, array)
```

```python
def remove_member (array, value):
    new_array = [item for item in array if item != value]
    return new_array # Hàm TRẢ VỀ kết quả
```

```python
array = [1,4,5,6,9]

remove_member(array, 5)
```
