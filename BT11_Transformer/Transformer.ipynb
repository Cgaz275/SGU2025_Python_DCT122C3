### =========Ví dụ 1: Dịch tiếng Anh → tiếng Việt============

```python
from transformers import pipeline
import transformers
transformers.utils.hub.cached_file = lambda *args, **kwargs: None  # xóa cache tạm thời
# Tạo pipeline dịch thuật
translator = pipeline("translation_en_to_vi", model="Helsinki-NLP/opus-mt-en-vi")

# Dùng
text = "Hello, how are you today?"
result = translator(text)
print(result[0]['translation_text'])
# Kết quả: Xin chào, bạn hôm nay thế nào?
```

### =========Ví dụ 2: Tạo văn bản tiếp theo (Text Generation) bằng GPT-2 nhỏ============

```python
# Ví dụ 2: Dùng PhoBERT để điền từ (masked language modeling) – rất vui
from transformers import pipeline

unmasker = pipeline("fill-mask", model="vinai/phobert-base")
result = unmasker("Hôm nay là một ngày rất <mask> đẹp.")
print(result[0]['sequence'])
# Kết quả ví dụ: Hôm nay là một ngày rất thật đẹp.
```
